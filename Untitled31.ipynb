```python
import streamlit as st
from transformers import pipeline
```

```python
st.title("Chatbot")
user_input = st.text_input("You:", "")
chat_history = st.text_area("Chat History:", "", height=200)
```

```python
chatbot = pipeline("text-generation", model="gpt2")

if st.button("Send"):
    chat_history += f"You: {user_input}\n"
    bot_response = chatbot(user_input)[0]['generated_text']
    chat_history += f"Chatbot: {bot_response}\n\n"

    st.text_area("Chat History:", chat_history, height=200)
```

```python
%%writefile chatbot_streamlit.py

import streamlit as st
from transformers import pipeline

def main():
    st.title("Chatbot")
    user_input = st.text_input("You:", "")
    chat_history = st.text_area("Chat History:", "", height=200)

    chatbot = pipeline("text-generation", model="gpt2")

    if st.button("Send"):
        if user_input.strip() == "":
            st.warning("Please enter a message.")
        else:
            chat_history += f"You: {user_input}\n"
            bot_response = chatbot(user_input)[0]['generated_text']
            chat_history += f"Chatbot: {bot_response}\n\n"

            st.text_area("Chat History:", chat_history, height=200)

if __name__ == "__main__":
    main()
```
